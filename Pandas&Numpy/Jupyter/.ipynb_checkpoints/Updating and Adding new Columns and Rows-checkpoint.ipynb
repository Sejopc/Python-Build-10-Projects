```python
import pandas
```

```python
df7 = pandas.read_json("http://pythonhow.com/supermarkets.json")
df7 = df7.set_index("Address")
df7
```

Adding a new Column

```python
df7["Continent"] = ["North America"] # it will give an error because there's only 1 item (and there are 6 rows)
```

```python
df7.shape[0] # number of rows
```

```python
df7.shape # number of rows and columns, respectively
```

```python
df7["Continent"] = df7.shape[0]*["North America"]
```

Modifying a column

```python
df7["Continent"] = df7["Country"] + ", " + "North America"
```

Adding a new Row

```python
df7_t = df7.T #it transposes the table rows and columns
```

```python
df7_t
```

```python
#Now that is transposed, we can add a new row as if it were a column (and then transpose it back; this as there is no
#method to add a row, so we have to kinda hack it).

df7_t["My new address"] = ["My City", "My Country", 10, 7, "My Shop", "My State", "My Continent"] #make sure to follow the columns (or rows in this case) order: address, city, country ...
```

```python
df7_t
```

```python
df7 = df7_t.T
```
