```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("sales-funnel(for_testing).xlsx")
df.head()
```

```python
#df["Status"] = df["Status"].astype("category")
#df["Status"].cat.set_categories(["won","pending","presented","declined"], inplace=True)
```

```python
pd.pivot_table(df,index=["Name"])
```

```python
pd.pivot_table(df,index=["Name", "Rep", "Manager"])
```

```python
pd.pivot_table(df,index=["Manager", "Rep"])
```

```python
pv = df.pivot_table(index=["Manager", "Rep"],values=["Price"])
print(pv)
```

```python
pv = df.pivot_table(index=["Manager", "Rep"],values=["Price"],aggfunc=np.sum)
print(pv)
```

```python
pv = df.pivot_table(index=["Manager", "Rep"],values=["Price"],aggfunc=np.mean)
print(pv)
```

```python
pv = df.pivot_table(index=["Manager", "Rep"],values=["Price"],aggfunc=[np.mean,len])
print(pv)
```

```python
pv = df.pivot_table(index=["Manager", "Rep"],values=["Price"],columns=["Product"],aggfunc=[np.sum])
print(pv)
```

```python
pv = df.pivot_table(index=["Manager", "Rep"],values=["Price"],columns=["Product"],aggfunc=[np.sum],fill_value=0)
print(pv)
```

```python
pv = df.pivot_table(index=["Manager", "Rep"],values=["Price", "Quantity"],columns=["Product"],aggfunc=[np.sum],fill_value=0)
print(pv)
```

```python
pv = df.pivot_table(index=["Manager", "Rep", "Product"],values=["Price", "Quantity"],aggfunc=[np.sum],fill_value=0)
print(pv)
```

```python
pv = df.pivot_table(index=["Manager", "Rep", "Product"],values=["Price", "Quantity"],aggfunc=[np.sum,np.mean],fill_value=0, margins=True)
print(pv)
```

```python
pv = df.pivot_table(index=["Manager","Status"])
print(pv)
```

```python
pv = df.pivot_table(index=["Manager","Status"], values=["Price"])
print(pv)
```

```python
pv = df.pivot_table(index=["Manager","Status"], values=["Price"],aggfunc=[np.sum], fill_value=0, margins=True)
print(pv)
```

```python
pv = df.pivot_table(index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
print(pv)
```
