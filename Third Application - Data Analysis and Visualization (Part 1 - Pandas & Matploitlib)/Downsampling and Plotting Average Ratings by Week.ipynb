```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv("reviews.csv", parse_dates=['Timestamp'])
```

### Rating average/count by week

```python
data
```

```python
data['Week'] = data['Timestamp'].dt.week
```

```python
data.head()
```

```python
data['Week'].max()
```

```python
data['Week'].min()
```

```python
# We can also filter by week yearly.
data['Week'] = data['Timestamp'].dt.strftime('%Y-%U')
data['Week']
```

```python
# Also
data['Week'] = data['Timestamp'].dt.strftime('%m %U') # Month and week of the year
data['Week']
```

```python
# However, we will use this one:
data['Week'] = data['Timestamp'].dt.strftime('%Y-%U')
week_average = data.groupby(['Week']).mean() # Average of rating by the weeks of year.
week_average
```

```python
week_average.index
```

```python
week_average['Rating']
```

```python
plt.figure(figsize=(25,3))
plt.plot(week_average.index, week_average['Rating'])
```
