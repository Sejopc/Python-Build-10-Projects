```python
import pandas
df = pandas.read_csv("supermarkets.csv")
df
```

```python
df["Address"] = df["Address"] + ", " + df["City"] + ", " + df["State"] + ", " + df["Country"]
```

```python
df
```

```python
df["Coordinates"] = df["Address"].apply(nom.geocode)
df
```

```python
df["Coordinates"][0]
```

```python
df["Coordinates"][0].latitude
```

```python
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude if x != None else None) 
# We cannot apply "latitude" method directly to df["Coordinates"] in this case, so we have to do it via apply method.
# What we are doing is passing every value of "Coordinates" row, to 'x' variable in Lambda. Then, we will put the 
# Latitude in our new column rows (Latitude), only if 'x' is not None (if it has lat and lon), otherwise, we will fill
# the row with 'None'. Pretty easy actually.
```

```python
df
```

```python
df["Longitude"] = df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
```
