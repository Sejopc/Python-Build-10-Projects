```python
import pandas
data = pandas.read_csv("reviews.csv")
```

## Time-based Filtering

```python
from datetime import datetime
```

```python
data[(data['Timestamp'] >= datetime(2020, 7, 1)) & (data['Timestamp'] <= datetime(2020, 12, 31))]

# This will produce error because we cannot use "<, >, <=, >=" comparisson operators between a str (data['Timestamp'])
# and a datetime. For this to work, we need to have both of same type (datetimes). For this, we need to add 
# parse_dates argument when reading the CSV file, as shown next.
```

```python
data2 = pandas.read_csv("reviews.csv", parse_dates=['Timestamp']) # we tell which column should be treated as datetimes
```

```python
data2[(data2['Timestamp'] >= datetime(2020, 7, 1)) & (data2['Timestamp'] <= datetime(2020, 12, 31))]
# Will still receive an error because these are 2 different datetypes objects (tz-naive and tz-aware)

```

```python
data2[(data2['Timestamp'] >= datetime(2020, 7, 1)) & (data2['Timestamp'] <= datetime(2020, 12, 31))]

```

```python
from pytz import utc
```

```python
data2[(data2['Timestamp'] >= datetime(2020, 7, 1, tzinfo=utc)) & (data2['Timestamp'] <= datetime(2020, 12, 31, tzinfo=utc))]

```
