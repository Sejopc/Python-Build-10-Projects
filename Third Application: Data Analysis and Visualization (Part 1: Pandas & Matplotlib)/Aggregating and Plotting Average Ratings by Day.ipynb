```python
import pandas
from datetime import datetime
from pytz import utc

```

```python
data = pandas.read_csv("reviews.csv", parse_dates=['Timestamp'])
```

```python
data.head()
```

```python
data['Day'] = data['Timestamp'].dt.date # Create a new Column name "Day" with only the date
data.head()
# day_average = data.groupby(['Timestamp'])
# day_average.head()
```

```python
data['Day'] = data['Timestamp'].dt.month # Create a new Column name "Day" with only the month
data.head()
```

```python
data['Day'] = data['Timestamp'].dt.date
day_average = data.groupby(['Day']).mean() # Get the mean of Ratings per day
day_average

# NOTE: In this case, there is only ONE column (Rating). "Day" is actually the Index, not a column.
# NOTE: The reason ONLY "Rating" column will show up and the others dissappear, is because the "mean()" method
# only works with Numberic column values.
```

```python
data['Day'] = data['Timestamp'].dt.date
day_average2 = data.groupby(['Day']).count() # Get the count of Ratings per day
day_average2.head()
```
