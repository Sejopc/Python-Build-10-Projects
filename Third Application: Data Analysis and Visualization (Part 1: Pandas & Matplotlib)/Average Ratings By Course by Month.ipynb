```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv("reviews.csv", parse_dates=['Timestamp'])

```

### Average Rating by Month by Course

```python
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')
```

```python
data.head()
```

```python
month_average = data.groupby(['Month']).mean()
```

```python
month_average.head() # This dataframe is missing the course names
```

```python
month_average_crs = data.groupby(['Month', 'Course Name']).mean()
month_average_crs
```

```python
# This Dataframe has 2 levels of Index. The Month and Course Name
month_average_crs.index
```

```python
month_average_crs.columns
```

```python
month_average_crs[:20]
```

```python
month_average_crs_unstack = data.groupby(['Month', 'Course Name']).mean().unstack()
month_average_crs_unstack[:20]
```

```python
month_average_crs_unstack.plot(figsize=(25,8))
```

```python
month_average_crs_unstack_count = data.groupby(['Month', 'Course Name']).count().unstack()
month_average_crs_unstack_count.plot(figsize=(25,8))
```
